```python
import numpy as np
import pandas as pd
import os
```

```python
permissions_df = pd.read_csv('../catch-o-extractor/extracted_permissions.csv')
labeled_data_df = pd.read_csv('./api_calls_labeled_dataset.csv')

labeled_data_df = labeled_data_df.drop('api_calls', axis=1)

permissions_df = permissions_df.drop(permissions_df.columns[1], axis=1)
permissions_df = permissions_df.rename(columns={permissions_df.columns[0]: 'name'})

df = permissions_df.merge(labeled_data_df, left_on='name', right_on='name', how='left')
df
```

```python
df.to_csv('permissions_labeled_dataset.csv', index=False)
```
